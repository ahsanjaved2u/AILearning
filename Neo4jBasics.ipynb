```python
from neo4j import GraphDatabase

class Neo4jRetriever:

    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def _fetch_nodes(tx):
        query = "MATCH (n:Person) RETURN n"
        result = tx.run(query)
        return result

    def close(self):
        self.driver.close()

    def get_nodes(self):
        with self.driver.session() as session:
            result = session.read_transaction(self._fetch_nodes)
            for record in result:
                print(record["n"])


    def get_node_properties(tx):
        query = "MATCH (n) RETURN n LIMIT 5"  # The Cypher query with alias 'n'
        result = tx.run(query)  # Run the query within the transaction context
        nodes = []
        for record in result:
            node = record["n"]  # Access the node using the alias 'n'
            properties = dict(node)  # Convert the Node object to a dictionary
            nodes.append(properties)  # Collect the properties of the node
        return nodes  # Return the list of node properties


# Usage
if __name__ == "__main__":
    uri = "bolt://localhost:7687"
    user = "neo4j"
    password = "password"

    retriever = Neo4jRetriever(uri, user, password)
    retriever.get_nodes()
    retriever.close()
```
